## **docker-compose.yml**

docker-compose.yml es el archivo que define y orquesta múltiples contenedores Docker como un solo sistema.

**docker-compose.yml es un archivo declarativo (YAML) donde defines:**

- qué servicios existen
- cómo se construyen
- cómo se conectan
- qué puertos usan
- qué variables de entorno tienen

Docker Compose luego levanta todo con un solo comando.

## **Sin Docker Compose tendrías que:**

1. correr cada servicio manualmente
2. configurar puertos a mano
3. asegurarte de que se vean entre sí
4. manejar redes manualmente

## **Con docker-compose.yml haces:**

            docker-compose up

**Y automáticamente:**

✔ levanta los 3 servicios

✔ crea una red interna

✔ conecta los servicios entre sí

✔ asigna puertos

✔ maneja dependencias

## **📦 Qué contiene normalmente**

**Un docker-compose.yml típico tiene:**

**1. services:**

Define cada microservicio (Cada uno es un contenedor independiente)

**2. build**

            build: .

Le dice a Docker: "construye la imagen usando el Dockerfile"

**3. ports**

ports:
  - "8001:8001"

Significa:

        host:contenedor

Ejemplo:

        localhost:8001 → service

**4. depends_on**

Indica si un servicio depende de otro (el servicio indicado debe levantarse antes)

⚠️ IMPORTANTE:

No garantiza que el servicio esté listo, solo que el contenedor arrancó.

**5. environment**

Sirve para: configuración dinámica (muy importante en producción)

**6. networks**

Docker Compose crea automáticamente una red:

- mi_proyecto_default

Todos los servicios viven ahí.
