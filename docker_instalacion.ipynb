## **Guía paso a paso para instalar Docker en Windows**

# **1. Verificar requisitos**

Antes de instalar Docker Desktop debes confirmar:

**Requisitos mínimos**

- Windows 10 o 11 de 64 bits

- Virtualización activada en BIOS

- WSL2 instalado

**Verificar virtualización**

- Presiona Ctrl + Shift + Esc

- Ve a Rendimiento → CPU

- Debe decir:

Virtualization: Enabled

Si dice Disabled, tendrás que activarlo en la BIOS.


# **2. Instalar WSL2 (recomendado por Docker)**

Abre PowerShell como administrador y ejecuta:

wsl --install

**Esto instalará:**

- Windows Subsystem for Linux

- Kernel Linux

- Ubuntu por defecto

Luego reinicia la computadora.

# **3. Descargar Docker**

Ve a la página oficial:

https://www.docker.com/products/docker-desktop/

**Descarga:**

Docker Desktop for Windows

**El software se llama:**

Docker Desktop

# **4. Instalar Docker Desktop**

Ejecuta el instalador descargado.

Durante la instalación asegúrate de que estén activadas estas opciones:

✔ Use WSL 2 instead of Hyper-V

✔ Add shortcut to desktop

Luego presiona Install.

La instalación tarda aproximadamente 3-5 minutos.

# **5. Reiniciar el sistema**

Cuando el instalador lo solicite:

Restart

# **6. Abrir Docker**

Después del reinicio:

1. Abre Docker Desktop

2. Espera a que aparezca:

Docker Desktop is running

# **7. Verificar que Docker funciona**

Abre PowerShell o CMD y ejecuta:

            docker --version

Deberías ver algo como:

            Docker version 25.x.x

Ahora prueba con un contenedor real:

            docker run hello-world

Docker descargará una imagen y mostrará:

Hello from Docker!

- Esto confirma que funciona correctamente.

## **8. Comandos básicos de Docker**

Al empezar con microservicios vas a usar mucho estos:

Ver contenedores activos 
       
            docker ps

Ver imágenes descargadas

            docker images

Ejecutar un contenedor

            docker run nombre_imagen

✅ Resultado final

Después de estos pasos tendrás:

- Docker funcionando

- entorno listo para microservicios

- posibilidad de crear Dockerfile y docker-compose
