## **Para instalar Windows Subsystem for Linux.**

ejecutar wsl --install 

y reiniciar, Windows abre automáticamente una terminal de Linux (normalmente Ubuntu) para terminar la configuración inicial. Si la ventana apareció unos segundos y se cerró, pueden haber pasado dos cosas comunes:

1️⃣ La configuración terminó automáticamente

En algunos casos WSL termina la instalación sin pedir interacción. Esto no es un problema.

2️⃣ Se cerró antes de completar la configuración

A veces la ventana se cierra antes de que configures el usuario de Linux.

## **Verificar si WSL quedó bien instalado**

Abre PowerShell o CMD y ejecuta:

            wsl --status

Deberías ver algo parecido a:

            Default Distribution: Ubuntu
            Default Version: 2

Luego ejecuta:

            wsl -l -v

Debería aparecer algo así:

            NAME      STATE           VERSION
            Ubuntu    Stopped         2

## **Si Ubuntu está instalado**

Simplemente abre Ubuntu manualmente:

1. Presiona Windows

2. Busca Ubuntu

3. Ábrelo

La primera vez aparecerá algo como:

            Installing, this may take a few minutes...

Luego te pedirá:

            Enter new UNIX username:

y después:

            Enter new password:

Ese será tu usuario de Linux dentro de WSL.

## **Después de eso**

Cuando Ubuntu termine de configurarse, ya podrás instalar:

- Docker Desktop

que usará WSL2 como backend, que es la forma moderna de correr Docker en Windows.

## **Paso siguiente: instalar Ubuntu en WSL**

En PowerShell ejecuta:

            wsl --install -d Ubuntu

Esto instalará:

Ubuntu dentro de WSL.

La instalación puede tardar 1–3 minutos.

## **Luego pasará esto**

Aparecerá una terminal que dirá algo como:

Installing, this may take a few minutes...

Después te pedirá crear usuario:

Enter new UNIX username:

Ejemplo:

user o admin

Luego contraseña:

Enter new password:

⚠️ Mientras escribes la contraseña no se verá nada en pantalla (es normal en Linux).

## **Verificar que quedó bien**

Después ejecuta:

            wsl -l -v

Deberías ver algo como:

            NAME      STATE           VERSION
            
            Ubuntu    Stopped         2

## **Cuando eso funcione**

Ya podrás instalar Docker Desktop, que utilizará WSL2 como motor.

## **Siguiente paso: instalar Docker**

Ahora puedes instalar:

- Docker Desktop

Descárgalo aquí:

https://www.docker.com/products/docker-desktop/

Durante la instalación asegúrate de que esté activado:

- Use WSL 2 instead of Hyper-V

Docker usará directamente tu Ubuntu de WSL como backend.

## **Después de instalar Docker**

La prueba final será ejecutar:

            docker run hello-world

Si aparece:

            Hello from Docker!

entonces tu entorno de contenedores está listo.

## **El error más común que ocurre justo después de instalar Docker con WSL **

(le pasa a casi todos la primera vez) y cómo evitarlo en 30 segundos.


Cuando ejecutas:

            docker run hello-world

aparece algo como:

            Cannot connect to the Docker daemon

o

            docker: command not found

Esto ocurre porque Docker aún no está integrado con tu Ubuntu de WSL.

## **Cómo evitar este problema (configuración correcta)**

Después de instalar Docker Desktop, haz lo siguiente:

**1. Abrir Docker Desktop**

Búscalo en el menú inicio y ábrelo.

Espera hasta que diga algo como:

Docker Desktop is running



**2. Activar integración con WSL**

Dentro de Docker Desktop:

1. Ir a Settings

2. Ir a Resources

3. Entrar en WSL Integration

Allí verás algo como:

            Enable integration with additional distros

Debes activar la distribución:

✔ Ubuntu

Esto conecta Docker con tu Linux.

**3. Aplicar cambios**

Presiona:

            Apply & Restart

Docker reiniciará su motor.

**Verificación final**

Ahora abre nuevamente tu terminal de Ubuntu y ejecuta:

            docker --version

Debería mostrar algo como:

            Docker version 26.x.x

Luego prueba:

            docker run hello-world

Si todo está bien verás:

            Hello from Docker!

## **Otro error muy frecuente (el segundo más común)**

Si aparece:

            WSL kernel version too old

solución:

            wsl --update

y luego:

            wsl --shutdown

Después abre Docker otra vez.
